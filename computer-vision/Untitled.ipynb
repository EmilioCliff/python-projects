```python
import cv2
from IPython.display import Image
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
Image('./piano2.jpg')
```

```python
img = cv2.imread('./piano2.jpg', cv2.IMREAD_COLOR)
print(img)
# plt.imshow(img[:,:, ::-1], cmap='gray')
```

```python
b, g, r = cv2.split(img)
plt.figure(figsize=[20, 8])
plt.subplot(141);plt.imshow(r, cmap="gray");plt.title('Red')
plt.subplot(142);plt.imshow(g, cmap="gray");plt.title('Green')
plt.subplot(143);plt.imshow(b, cmap="gray");plt.title('Blue')
merged = cv2.merge((b, g, r))
plt.subplot(144);plt.imshow(merged[:,:,::-1], cmap="gray");plt.title('Merged')
```

```python
new_img = img.copy()
# new_imgrgb = new_img[:,:,::-1]
cv2.circle(new_img, (200,100), 100, color=(0,0,255), thickness=2)
plt.imshow(new_img[:,:,::-1])
```

```python
img2_bgr = cv2.imread('./piano2.jpg')
cv2.rectangle(img2_bgr, (100,10), (250,400), thickness=2, color=(0,255,0))
plt.imshow(img2_bgr[:,:,::-1])
```

```python
flipped_img = cv2.flip(img2_bgr, 0)
cv2.putText(flipped_img, text='Momma',org=(100,300), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1, color=(255,0,0),thickness=2)
plt.imshow(flipped_img)
```
